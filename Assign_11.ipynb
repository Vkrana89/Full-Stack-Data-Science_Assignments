1. Create an assert statement that throws an AssertionError if the variable spam is a negative
integer.

The assert statement can be used to check a condition and raise an AssertionError if the condition is False. To throw an AssertionError if the variable spam is a negative integer, you can use the following assert statement:

assert spam >= 0, "spam must not be a negative integer"

If the value of spam is a negative integer, this assert statement will raise an AssertionError with the specified error message.

4. What are the two lines that must be present in your software in order to call logging.debug()?
In order to call logging.debug(), you need to have the following two lines in your software to configure the logging system:

import logging
logging.basicConfig(level=logging.DEBUG)



5. What are the two lines that your program must have in order to have logging.debug() send a logging message to a file named programLog.txt?
To have logging.debug() send a logging message to a file named programLog.txt, you need the following two lines along with the previous logging configuration:

import logging
logging.basicConfig(filename='programLog.txt', level=logging.DEBUG)

The filename argument in the basicConfig() function specifies the file to which the logging messages will be written.

6. What are the five levels of logging?
The five levels of logging in increasing order of severity are:
DEBUG: Detailed information, typically used for debugging purposes.
INFO: General information about the program's execution.
WARNING: Indicates a potential problem or an issue that may need attention.
ERROR: Indicates a more serious problem or error that may cause the program to malfunction.
CRITICAL: Indicates a critical error that may result in the program's failure.

7. What line of code would you add to your software to disable all logging messages?
To disable all logging messages, you can add the following line of code to your software:
logging.disable(logging.CRITICAL)
This line sets the logging level to CRITICAL, which effectively disables all logging messages with lower severity levels.


8. Why is using logging messages better than using print() to display the same message?
Using logging messages is generally considered better than using print() to display the same message because:
Logging messages can be easily controlled and configured by setting the log level. You can control which messages are shown based on their severity level, making it easier to filter or silence specific types of messages.
Logging messages can be redirected to various outputs, such as the console, files, or external services, without modifying the code. This provides flexibility in handling and storing log information.
Logging messages can include additional information like timestamps, log levels, and source module names, which can be useful for debugging and troubleshooting.
Logging messages can be enabled or disabled globally for an application, allowing for easy debugging or production use without modifying the code.

9. What are the differences between the Step Over, Step In, and Step Out buttons in the debugger?
In the context of a debugger:
Step Over: Executes the current line of code and moves to the next line. If the current line contains a function call, it will be executed entirely, and the debugger will move to the next line after the function call returns.
Step In: If the current line of code contains a function call, the debugger will enter the function and move to the first line of the called function, allowing you to debug the function's execution.
Step Out: If the debugger is currently inside a function, the Step Out command will execute the remaining lines of the current function and return to the calling function. It allows you to quickly skip the detailed debugging of a function and move up the call stack.

10. After you click Continue, when will the debugger stop ?
After clicking Continue, the debugger will stop when it encounters the next breakpoint or when the program execution completes.

11. What is the concept of a breakpoint?
 A breakpoint is a designated point in the code where the debugger will pause program execution, allowing you to inspect variables, step through code, and analyze the program's state at that specific point. It is a useful tool for debugging as it allows you to examine and control the program's execution flow.
